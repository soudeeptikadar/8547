```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
path='/content/drive/MyDrive/FOR BATA csv.csv'
shoesdata = pd.read_csv(path)
shoesdata.head()
```

```python
shoesdata['Brand_Name'].value_counts()[0:5]
```

```python
plt.figure(figsize=(10,5))
plt.bar(shoesdata['Brand_Name'].value_counts()[0:5].index,shoesdata['Brand_Name'].value_counts()[0:5].values)
plt.show()
```

```python
shoesdata['RATING'].value_counts()
```

```python
plt.figure(figsize=(10,5))
plt.pie(shoesdata['RATING'].value_counts(),labels=shoesdata['RATING'].value_counts().index,autopct='%1.2f%%')
plt.show()
```

```python
bata_high_rating = shoesdata[(shoesdata['Brand_Name'] == 'BATA') & (shoesdata['RATING'] > 3.5)]
count_bata_high_rating = bata_high_rating.shape[0]

print(f"Number of Bata shoes with a rating above 3.5: {count_bata_high_rating}")
```

```python
bata_high_rating = shoesdata[(shoesdata['Brand_Name'] == 'BATA') & (shoesdata['RATING'] > 4.0)]
count_bata_high_rating = bata_high_rating.shape[0]

print(f"Number of Bata shoes with a rating above 4.0: {count_bata_high_rating}")
```
